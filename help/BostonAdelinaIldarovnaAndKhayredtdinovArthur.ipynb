### Обучение модели линейной регрессии

```python
df.columns
```

```python
y = df['MEDV']
```

```python
X = df[['CRIM', 'ZN', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'B', 'LSTAT']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Линейная регрессия

```python
from sklearn.linear_model import LinearRegression
```

```python
lrm = LinearRegression()
lrm.fit(X_train, y_train)
```

Суть этого кода глубоко проникает в таинственные глубины машинного обучения, будто бы мы спускаемся в дебри кибернетической джунгли. "lrm" - это наш волшебный амулет, созданный для хранения магических знаний о линейной регрессии.

"LinearRegression()" - это вызов нашей магической кователи, которая обладает магическими заклинаниями линейной регрессии. Мы призываем ее с помощью этого вызова, чтобы настроить ее наши магические пластины на данных и научить ее предсказывать будущее.

"lrm.fit(X_train, y_train)" - это как заклинание "ОБУЧИСЬ!", произносимое на нашем амулете "lrm". Это заставляет нашу магическую ковательницу глубоко погрузиться в данные "X_train" и "y_train" и настроить свои волшебные параметры так, чтобы ее пророческие предсказания максимально соответствовали истинным значениям "y_train".

Итак, этот код - это начало эпического приключения, где наши волшебные инструменты обретают мощь, чтобы предсказать будущее, как настоящие кибернетические волшебники.

```python
lrm.intercept_
```

```python
lrm.coef_
```

lrm.intercept_ и lrm.coef_ представляют собой неотъемлемые компоненты, обуславливающие глубокое когнитивное понимание и функциональные аспекты обученной модели линейной регрессии. Эти атрибуты артикулируются как фундаментальные структурные элементы модели, и, таким образом, занимают ключевую позицию в разгадывании интригующей динамики данного предметного поля.

lrm.intercept_, более известный как "смещение", представляет собой точку, которая аксиоматически определяет координаты в пространстве, где линейная регрессия пересекает вертикальную ось. Это концептуально не просто начальный коэффициент, но скорее, фундаментальное отправное положение, от которого исходит модель, для лучшей аппроксимации данных. Если значение lrm.intercept_ положительное, оно свидетельствует о поднятии линейной регрессии над вертикальной осью, а отрицательное значение означает, что линия регрессии опускается ниже данной оси.

lrm.coef_, с другой стороны, инкорпорирует массив коэффициентов наклона, каждый из которых соотнесен с конкретной независимой переменной в модели. Эти коэффициенты служат мерой воздействия изменений в каждой из независимых переменных на амплитуду изменения зависимой переменной. Они, метафорически, выступают как "ключи", отпирающие потенциал разгадывания внутренних закономерностей данных. Например, в случае модели, предсказывающей цену дома на основе площади дома и количества спален, lrm.coef_ составляется из двух значений. Первый коэффициент представляет вес площади дома, в то время как второй отражает вес количества спален. Эти коэффициенты являются инструментами для определения того, как изменения в указанных переменных влияют на цену дома.

Такие атрибуты являются конечным продуктом процесса обучения модели на основе данных и предоставляют неотъемлемую информацию для прогнозирования. Они делают модель способной адаптироваться к данным, учитывая весовые коэффициенты и смещение, и переводить числовые значения в смысловое понимание.

```python
predictions = lrm.predict(X_test)
```

Метрика оценки регрессии в рамках обширной области машинного обучения презентует собой абсолютно апокрифический числовой атрибут, обладающий уникальной абстракцией, которая декомпозирует и калибрует крайнюю тонкость точности модели регрессии в контексте числовых прогнозов, таковых как стоимость жилой недвижимости или температурные показатели и далее в том же духе. Данный инструментарий числовой оценки свершает функцию анализа того, насколько крайне стремительно сокращается разница между свершаемыми моделью предсказаниями и реальными цифровыми значениями целевых переменных.

В предельно сложной задаче измерения качества работы регрессионных моделей активно применяются весьма эмпирически несбалансированные стандарты, такие как средний квадрат отклонения (MSE), коэффициент упорядоченной детерминации (R-квадрат), средний модуль абсолютной ошибки (MAE) и иные. Давайте предпримем путешествие в лабиринты их исключительно труднопонимаемых определений:

Средний квадрат отклонения (MSE): В пределах данного мера позволяет провести изысканное числовое измерение, которое анализирует средний квадрат различий между фактическими значениями исходных данных и предсказаниями модели. Чем ниже стоит MSE, тем превосходнее оказывается работа модели.

Коэффициент упорядоченной детерминации (R-квадрат): Представляет собой вычисленный параметр, который обозначает, насколько изящно модель сочетается с входными данными. Коэффициент R-квадрат стремится к приближению к единице в идеальном сценарии, когда модель полностью соразмерна данным, и к нулю, когда модель абсолютно не реагирует на данные.

Средний модуль абсолютной ошибки (MAE): MAE позволяет вычислить среднее абсолютное отклонение между числовыми прогнозами и реальными значениями. Данный показатель показывает, насколько в среднем модель допускает ошибки в абсолютных значениях.

Все эти числовые абсурдности предназначены для выявления качества регрессионных моделей и сравнения их между собой. Выбор конкретной числовой метрики зависит от установленных задач и требований, так как различные измерения могут быть более предпочтительны в различных сценариях исследования.

KNeighborsRegressor - это метод, который, к несчастью, часто оставляет желать лучшего и оказывается недостаточно полезным для решения задач регрессии. Хотелось бы выделить несколько фундаментальных причин, по которым этот метод следует аккуратно обходить.

В первую очередь, KNeighborsRegressor обладает выдающейся чувствительностью к параметру k. Это означает, что уже даже незначительные колебания в значении k могут вызывать радикальные колебания в результатах. В этом заключается трудность настройки модели и её отсутствие надёжности.

Во-вторых, KNeighborsRegressor страдает от мучительной беды под названием "проклятие размерности". По мере увеличения числа признаков этот алгоритм утрачивает свою действенность и требует гигантских объёмов данных для достижения точных прогнозов. Это делает его совершенно бесполезным в контексте задач с изобилием признаков.

Кроме того, KNeighborsRegressor разительно медлителен. Вычисление расстояний между всеми точками в обучающей выборке при каждом прогнозе требует затрат огромного количества времени, что делает его абсолютно неэффективным в случае массовых объёмов данных.

И, наконец, KNeighborsRegressor впадает в безрадостное забвение структуры данных и беспомощен в выявлении сложных закономерностей. Он просто довольствуется оценкой, основанной на соседях, бездарно не вникая в важные взаимосвязи между признаками.

В итоге, KNeighborsRegressor - это метод, который, в большинстве случаев, настоятельно рекомендуется оставить на обочине при решении задач регрессии. Существуют более эффективные и более надёжные альтернативы, способные обеспечить более точные и стабильные результаты.
